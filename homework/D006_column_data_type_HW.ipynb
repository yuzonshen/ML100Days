```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path, 並讀取 app_train
dir_data = './data/'
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

## 作業
將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train.head()
```

```python
print(app_train.shape)
app_train.head()
app_train['WEEKDAY_APPR_PROCESS_START'].value_counts()
```

```python
#One Hot encoding方法

# 先將需要的欄取出
#sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
sub_train = app_train[['WEEKDAY_APPR_PROCESS_START']]

#columns 代表取出欄位,透過 prefix 來調整欄位名稱 (預設: 原欄位名加上_value值)
#dummy=pd.get_dummies(sub_train, columns=['WEEKDAY_APPR_PROCESS_START'], prefix=['sub'])
dummy=pd.get_dummies(sub_train, columns=['WEEKDAY_APPR_PROCESS_START'])

#資料合併 : axis=1 (行合併)
data=pd.concat([app_train,dummy],axis=1)


print(data.shape)
data
#最後邊多7欄: One Hot encoding 產生的dummy欄位
```
