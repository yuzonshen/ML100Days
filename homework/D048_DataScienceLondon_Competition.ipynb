```python
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 設定 data_path
dir_data = './data/'
f_app_train = os.path.join(dir_data, 'DS_train.csv')
f_app_train_label = os.path.join(dir_data, 'DS_trainLabels.csv')
f_app_test = os.path.join(dir_data, 'DS_test.csv')


# 讀取檔案
app_train = pd.read_csv(f_app_train, header=None)
app_train_label = pd.read_csv(f_app_train_label, header=None)
app_test = pd.read_csv(f_app_test, header=None)
```

```python
app_train
```

```python
print(app_train.shape)
print(app_train_label.shape)
print(app_test.shape)
```

```python
app_train.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# 縮放器 : 設定特徵縮放到 0~1 區間
scaler = MinMaxScaler(feature_range = (0, 1))

# 縮放器載入 train 的上下限, 對 train, test 進行縮放轉換
scaler.fit(app_train)
train = scaler.transform(app_train)
test = scaler.transform(app_test)
print(train.shape)
print(test.shape)
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(train, np.ravel(app_train_label), test_size=0.25)
print(x_train.shape)
print(y_train.shape)
```

```python
# Model: 隨機森林
# n_estimators=10, #決策樹的數量
# max_features="auto", #選取的特徵數量 The number of features to consider when looking for the best split: 
#                      If “auto”, then max_features=sqrt(n_features)
# 建立模型 (使用 20 顆樹，每棵樹的最大深度為 4)
clf = RandomForestClassifier(n_estimators=50, max_depth=4, max_features='auto')

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
y_pred
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```

```python
#multiclass多分類評分: 須加上average參數 average{'micro', 'macro', 'samples','weighted', 'binary'} or None, default='binary'
f1 = metrics.f1_score(y_test, y_pred) # 使用 F1-Score 評估: If average=None, the scores for each class are returned.
precision = metrics.precision_score(y_test, y_pred) # 使用 Precision 評估
recall  = metrics.recall_score(y_test, y_pred) # 使用 recall 評估
print("F1-Score: ", f1) 
print("Precision: ", precision)
print("Recall: ", recall)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Model: 梯度提升機
# 建立模型
clf = GradientBoostingClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
y_pred
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```

```python
#multiclass多分類評分: 須加上average參數 average{'micro', 'macro', 'samples','weighted', 'binary'} or None, default='binary'
f1 = metrics.f1_score(y_test, y_pred) # 使用 F1-Score 評估: If average=None, the scores for each class are returned.
precision = metrics.precision_score(y_test, y_pred) # 使用 Precision 評估
recall  = metrics.recall_score(y_test, y_pred) # 使用 recall 評估
print("F1-Score: ", f1) 
print("Precision: ", precision)
print("Recall: ", recall)
```

```python
#以Model: 梯度提升機 訓練 all trainset, 預測testset
# 建立模型
clf = GradientBoostingClassifier()

app_train_label = np.ravel(app_train_label)
# 訓練模型
clf.fit(train, app_train_label)

# 預測測試集
y_pred = clf.predict(test)
y_pred
```

```python
y_pred.shape
```

```python
# 計算提交結果
submit = pd.DataFrame(y_pred).reset_index()
submit.columns = ['Id','Solution']
submit['Id'] = submit['Id'] + 1
submit
```

```python
out_dir = './output/'
f_out = os.path.join(out_dir, 'DataScienceLondon_predict1.csv')
submit.to_csv(f_out, index=False)
```
